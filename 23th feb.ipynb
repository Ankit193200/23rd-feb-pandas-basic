# Q4:
A DataFrame in pandas is a 2-dimensional labeled data structure with columns that can be of different types. It can be thought of as an Excel spreadsheet or SQL table. In contrast, a Series is a 1-dimensional labeled array capable of holding any data type. A DataFrame is essentially a collection of Series objects.

# Q5:
Common functions to manipulate data in a Pandas DataFrame include head(), tail(), describe(), groupby(), sort_values(), filter(), pivot_table(), and many more.

Example:

python
Copy code
  # Using some common DataFrame manipulation functions
df_example = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
})

# Head
print(df_example.head())

# Describe
print(df_example.describe())

# Sorting
print(df_example.sort_values(by='B'))

# Groupby
grouped_data = df_example.groupby('A').sum()
print(grouped_data)
